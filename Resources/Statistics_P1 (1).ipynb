(1) Outlier detection and removal using Standard Deviation

```python
min_limit, max_limit = df.price.quantile([0.0001, 0.999])
min_limit, max_limit
```

```python
df[df.price<min_limit]
```

```python
outliers = df[(df.price>max_limit) | (df.price<min_limit)]
outliers
```
